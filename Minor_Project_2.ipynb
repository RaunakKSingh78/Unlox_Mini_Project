```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_excel("4870116-Student_Academic_Performance.xlsx")
print(df)
```

```python
print(f"The shape of the dataset is :- {df.shape}")
print(f"The columns of the dataset are :- \n{df.columns}\n")
print(f"The datatype of the dataset is :- \n{df.dtypes}")
```

```python
df.isnull().sum()
# because there is no null values we don't require to fix the null/empty parts of the dataset by using .fillna()
```

```python
df['State Of Birth'].value_counts()
```

```python
#Histogram of CGPA Distribution also the same for marks distribution

plt.figure(figsize=(5,5))
plt.hist(df["CGPA"], bins=20)
plt.title("CGPA Distribution")
plt.xlabel("CGPA")
plt.ylabel("Frequency")
plt.show()
```

```python
#Also for attendance categories
plt.figure(figsize=(10,5))
sns.barplot(x="Announcements Visited" ,y="GPA", data=df)
plt.title("Bar plots for attendance categories vs GPA")
plt.show()
```

```python
#Also for subject-wise performance
plt.figure(figsize=(15,5))
sns.boxplot(x="Course" ,y="GPA", data=df)
plt.ylabel("GPA")
plt.show()
```
